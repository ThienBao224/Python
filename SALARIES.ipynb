# Phân tích dữ liệu Salaries

**1. Input**

Bảng dữ liệu Salaries.csv gồm các cột:

Id: mã định danh nhân viên.

EmployeeName: tên nhân viên.

JobTitle: chức danh công việc.

BasePay: lương cơ bản.

OvertimePay: lương làm thêm giờ.

OtherPay: các khoản phụ cấp khác.

Benefits: các khoản phúc lợi.

TotalPay: tổng lương

TotalPayBenefits: tổng lương kèm phúc lợi.

Year: năm thống kê.

Notes: ghi chú

Agency: cơ quan làm việc.

Status: trạng thái

Đọc file

```python
import pandas as pd

# Đọc file CSV từ Drive
df = pd.read_csv("/content/drive/MyDrive/6_Nguyễn Hoàng Thiên Bảo_Tuần4/Buổi 2 _ Thứ 7/Salaries.csv")

# Hiển thị 5 dòng đầu
df.head()
```

**Data preprocessing**

Kiểm tra xem dữ liệu có bị null hay không

```python
print(df.isnull().sum())
```

```python
# Dimension of the dataset
print(df.shape)
```

**Lọc dataframe**

```python
# Lọc nhân viên có lương cơ bản > 100,000
high_salary = df[df['BasePay'] > 100000]
display(high_salary)
```

```python
# Lọc nhiều điều kiện: lương > 100k và năm 2014
filtered = df[(df['BasePay'] > 100000) & (df['Year'] == 2014)]
display(filtered)
```

```python
# Lọc theo chức danh có chứa từ 'POLICE'
police = df[df['JobTitle'].str.contains('POLICE', case=False, na=False)]
display(police)
```

```python
# Lọc nhân viên làm việc năm 2014
df_2014 = df[df['Year'] == 2014]
display(df_2014)
```

Ta dùng hàm df.select_dtypes() để lọc các cột theo định dạng dữ liệu. Những định dạng chính bao gồm integer, float, object, boolean.

```python
# lọc các trường có định dạng dữ liệu là float.
df.select_dtypes('float').head()
```

Min, max, mean, meadian, sum

```python
cols = ['BasePay', 'OvertimePay', 'TotalPayBenefits']

stats = pd.DataFrame({
    'Min': df[cols].min(),
    'Max': df[cols].max(),
    'Mean': df[cols].mean(),
    'Median': df[cols].median(),
    'Sum': df[cols].sum()
})
print(stats)
```

```python
# Sắp xếp tăng dần theo BasePay
df_sorted = df.sort_values(by='BasePay', ascending=True)
display(df_sorted)
```

```python
# Sắp xếp giảm dần theo TotalPayBenefits
df_sorted_desc = df.sort_values(by='TotalPayBenefits', ascending=False)
display(df_sorted_desc)
```

**Thống kê theo nhóm**


Trung bình tổng lương theo năm

```python
salary_by_year = df.groupby('Year')['TotalPayBenefits'].mean()
print(salary_by_year)
```

Tổng lương theo chức danh

```python
salary_by_job = df.groupby('JobTitle')['TotalPayBenefits'].sum()
salary_by_job.head()
```

Nhiều phép thống kê cùng lúc

```python
stats_by_year = df.groupby('Year')['TotalPayBenefits'].agg(['min', 'max', 'mean', 'median', 'sum'])
print(stats_by_year)
```

**Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(8,5))
sns.boxplot(x='Year', y='TotalPayBenefits', data=df)
plt.title("So sánh tổng lương theo năm")
plt.xlabel("Năm")
plt.ylabel("TotalPayBenefits")
plt.show()
```

```python
yearly_salary = df.groupby('Year')['TotalPayBenefits'].mean()

plt.figure(figsize=(8,5))
sns.lineplot(x=yearly_salary.index, y=yearly_salary.values, marker='o')
plt.title("Xu hướng lương trung bình theo năm")
plt.xlabel("Năm")
plt.ylabel("Lương trung bình")
plt.show()
```

```python
# Cài đặt kích thước hình
sns.set(rc={'figure.figsize': (10, 8.27)})

# Vẽ histogram cho BasePay
sns.histplot(df['BasePay'], bins=50, kde=True)
plt.title("Phân phối lương cơ bản (BasePay)")
plt.xlabel("BasePay")
plt.ylabel("Số lượng nhân viên")
plt.show()
```

```python
sns.set(rc={'figure.figsize': (10, 8)})

sns.histplot(df['TotalPayBenefits'], bins=50, kde=True)
plt.title("Phân phối tổng lương + phúc lợi")
plt.xlabel("TotalPayBenefits")
plt.ylabel("Số lượng nhân viên")
plt.show()
```

**Correlation matrix**

```python
numeric_df = df.drop(columns=['Notes', 'Status']).select_dtypes(include=['float64','int64'])
correlation_matrix = numeric_df.corr().round(2)
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", fmt=".2f", square=True)
plt.xticks(rotation=45)
plt.show()
```

**Observations**

```python
features = ['BasePay', 'OvertimePay']
target = df['TotalPayBenefits']

plt.figure(figsize=(20, 5))

for i, col in enumerate(features):
    plt.subplot(1, len(features), i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='o', alpha=0.5)
    plt.title(f"{col} vs TotalPayBenefits")
    plt.xlabel(col)
    plt.ylabel("TotalPayBenefits")

plt.tight_layout()
plt.show()
```

```python
df_clean = df.drop(columns=['Notes', 'Status'], errors='ignore')
numeric_cols = df_clean.select_dtypes(include=["float64", "int64"]).columns

fig, axs = plt.subplots(ncols=4, nrows=3, figsize=(20, 10))
axs = axs.flatten()

index = 0
for col in numeric_cols:
    sns.histplot(df_clean[col], kde=True, ax=axs[index])
    axs[index].set_title(col, fontsize=10)
    axs[index].set_xlabel("")
    axs[index].set_ylabel("")
    index += 1

for j in range(index, len(axs)):
    fig.delaxes(axs[j])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
plt.show()
```
